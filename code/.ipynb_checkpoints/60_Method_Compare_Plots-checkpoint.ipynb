```python
# Jack
dj = pd.read_csv("../results/year_coefs_j.csv")
dj.columns = ["label","est","se","t","p"]
dj['year'] = dj['label'].str.extract("\(first_year\), \"1997\"\)(\d{4})").astype(float)
dj = dj.dropna()
Ej = dj.set_index(["year"])['est']
SEj = dj.set_index(["year"])['se']*1.96
Ej.loc[1997] = 0
SEj.loc[1997] = 0
Ej = Ej.sort_index()
SEj = SEj.sort_index()

m1 = pd.read_csv("../results/m1_year_j.csv")
m1.columns = ["label","est","se","t","p"]
m1['year'] = m1['label'].str.extract("\(first_year\), \"1997\"\)(\d{4})").astype(float)
m1 = m1.dropna()
Ej1 = m1.set_index(["year"])['est']
SEj1 = m1.set_index(["year"])['se']*1.96
Ej1.loc[1997] = 0
SEj1.loc[1997] = 0
Ej1 = Ej1.sort_index()
SEj1 = SEj1.sort_index()

mfe2 = pd.read_csv("../results/mfe2_year_j.csv")
mfe2.columns = ["label","est","se","t","p"]
mfe2['year'] = mfe2['label'].str.extract("\(first_year\), \"1997\"\)(\d{4})").astype(float)
mfe2 = mfe2.dropna()
FE2 = mfe2.set_index(["year"])['est']
FE2SE = mfe2.set_index(["year"])['se']*1.96
FE2.loc[1997] = 0
FE2SE.loc[1997] = 0
FE2 = FE2.sort_index()
FE2SE = FE2SE.sort_index()

mfe1 = pd.read_csv("../results/mfe1_year_j.csv")
mfe1.columns = ["label","est","se","t","p"]
mfe1['year'] = mfe1['label'].str.extract("\(first_year\), \"1997\"\)(\d{4})").astype(float)
mfe1 = mfe1.dropna()
FE1 = mfe1.set_index(["year"])['est']
FE1SE = mfe1.set_index(["year"])['se']*1.96
FE1.loc[1997] = 0
FE1SE.loc[1997] = 0
FE1 = FE1.sort_index()
FE1SE = FE1SE.sort_index()

mre1 = pd.read_csv("../results/mre1_year_2.csv")
mre1.columns = ["label","est","se","t","p"]
mre1['year'] = mre1['label'].str.extract("\(first_year\), \"1997\"\)(\d{4})").astype(float)
mre1 = mre1.dropna()
RE1 = mre1.set_index(["year"])['est']
RE1SE = mre1.set_index(["year"])['se']*1.96
RE1.loc[1997] = 0
RE1SE.loc[1997] = 0
RE1 = RE1.sort_index()
RE1SE = RE1SE.sort_index()

mre1 = pd.read_csv("../results/mre2_year_j.csv")
mre1.columns = ["label","est","se","t","p"]
mre1['year'] = mre1['label'].str.extract("\(first_year\), \"1997\"\)(\d{4})").astype(float)
mre1 = mre1.dropna()
RE2j = mre1.set_index(["year"])['est']
RE2jSE = mre1.set_index(["year"])['se']*1.96
RE2j.loc[1997] = 0
RE2jSE.loc[1997] = 0
RE2j = RE2j.sort_index()
RE2jSE = RE2jSE.sort_index()


# Williams 82
wj = pd.read_csv("../results/year_coefs_w.csv")
wj.columns = ["label","est","se","t","p"]
wj['year'] = wj['label'].str.extract("\(first_year\), \"1997\"\)(\d{4})").astype(float)
wj = wj.dropna()
Ew = wj.set_index(["year"])['est']
SEw = wj.set_index(["year"])['se']*1.96
Ew.loc[1997] = 0
SEw.loc[1997] = 0
Ew = Ew.sort_index()
SEw = SEw.sort_index()

m1w = pd.read_csv("../results/m1_year_w.csv")
m1w.columns = ["label","est","se","t","p"]
m1w['year'] = m1w['label'].str.extract("\(first_year\), \"1997\"\)(\d{4})").astype(float)
m1w = m1w.dropna()
Ew1 = m1w.set_index(["year"])['est']
SEw1 = m1w.set_index(["year"])['se']*1.96
Ew1.loc[1997] = 0
SEw1.loc[1997] = 0
Ew1 = Ew1.sort_index()
SEw1 = SEw1.sort_index()

mfe2 = pd.read_csv("../results/mfe2_year_w.csv")
mfe2.columns = ["label","est","se","t","p"]
mfe2['year'] = mfe2['label'].str.extract("\(first_year\), \"1997\"\)(\d{4})").astype(float)
mfe2 = mfe2.dropna()
FE2w = mfe2.set_index(["year"])['est']
FE2wSE = mfe2.set_index(["year"])['se']*1.96
FE2w.loc[1997] = 0
FE2wSE.loc[1997] = 0
FE2w = FE2w.sort_index()
FE2wSE = FE2wSE.sort_index()

mfe1 = pd.read_csv("../results/mfe1_year_w.csv")
mfe1.columns = ["label","est","se","t","p"]
mfe1['year'] = mfe1['label'].str.extract("\(first_year\), \"1997\"\)(\d{4})").astype(float)
mfe1 = mfe1.dropna()
FE1w = mfe1.set_index(["year"])['est']
FE1wSE = mfe1.set_index(["year"])['se']*1.96
FE1w.loc[1997] = 0
FE1wSE.loc[1997] = 0
FE1w = FE1w.sort_index()
FE1wSE = FE1wSE.sort_index()

mre1 = pd.read_csv("../results/mre1_year_3.csv")
mre1.columns = ["label","est","se","t","p"]
mre1['year'] = mre1['label'].str.extract("\(first_year\), \"1997\"\)(\d{4})").astype(float)
mre1 = mre1.dropna()
RE1w = mre1.set_index(["year"])['est']
RE1wSE = mre1.set_index(["year"])['se']*1.96
RE1w.loc[1997] = 0
RE1wSE.loc[1997] = 0
RE1w = RE1w.sort_index()
RE1wSE = RE1wSE.sort_index()

mre1 = pd.read_csv("../results/mre2_year_w.csv")
mre1.columns = ["label","est","se","t","p"]
mre1['year'] = mre1['label'].str.extract("\(first_year\), \"1997\"\)(\d{4})").astype(float)
mre1 = mre1.dropna()
RE2w = mre1.set_index(["year"])['est']
RE2wSE = mre1.set_index(["year"])['se']*1.96
RE2w.loc[1997] = 0
RE2wSE.loc[1997] = 0
RE2w = RE2w.sort_index()
RE2wSE = RE2wSE.sort_index()
```

```python
f,a = plt.subplots(figsize=(6,4))

offset = .2
Xrange1 = [int(x) +offset for x in Ej1.index]
Xrange2 = [int(x) -offset for x in FE2.index]
Xrange3 = [int(x) - 0 for x in Ej.index]


a.errorbar(Xrange3[2:],Ej.iloc[2:],yerr=SEj.iloc[2:],capsize=3,marker="o",ms=3,color="C0",label="Fixed Coefficient")
a.errorbar(Xrange1[2:],Ej1.iloc[2:],yerr=SEj1.iloc[2:],capsize=3,marker="o",ms=3,color="C0",ls="--",label="Varying Coefficient",alpha=.5)
# a.errorbar(Xrange2[2:],FE2.iloc[2:],yerr=FE2SE.iloc[2:],capsize=3,marker="o",ms=3,color="midnightblue",label="VFE Model, Check Yield Control",ls=":")

plt.legend()

plt.axhline(0,color="black",ls="--")
plt.xticks([2000,2005,2010,2015,2020])
plt.savefig("../figs/VaryingFixedCompare_MG2_1yr.pdf",dpi=300,bbox_inches="tight")
```
