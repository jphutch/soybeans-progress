```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d = pd.read_csv("../data/output_data/Full_Trial_Data.csv").drop_duplicates()
```

```python
d[~d['public_variety']][d.location!="region"].groupby(["year","variety_id"])['region'].nunique().unstack().T.mean().plot(color="black",label="Private")
d[d['public_variety']][d.location!="region"].groupby(["year","variety_id"])['region'].nunique().unstack().T.mean().plot(color="C1",label="Public")
avg = d[d.location!="region"].groupby(["year","variety_id"])['region'].nunique().unstack().T.mean().mean()
plt.axhline(avg,color="black",ls=":",label="Sample Average")

plt.xlabel("Trial Year")
plt.ylabel("Average Number of Region Entries per Variety")
plt.legend()
plt.ylim(0,5.5)
plt.savefig("../figs/Regions_Per_Year.pdf",dpi=300,bbox_inches="tight")
```

```python
matrix = d[d.location!="region"].groupby(['year'])['variety_id'].value_counts().unstack()
matrix_private = d[d.location!="region"][~d['public_variety']].groupby(['year'])['variety_id'].value_counts().unstack()
matrix_public = d[d.location!="region"][d['public_variety']].groupby(['year'])['variety_id'].value_counts().unstack()

particp = (~pd.isnull(matrix)).astype(int)

part_years =  particp.sum()
part_years = part_years.reset_index()
part_years.columns = ['variety_id','years_in_trial']
d = d.merge(part_years,how="left")
```

```python
particp.sum().to_csv("../data/output_data/Years_Tested.csv")
```

```python
d[d.location!="region"].groupby(['year'])['variety_id'].value_counts().mean()
```

```python
matrix_private.mean(axis=1).plot(label="Private",color="black")
matrix_public.mean(axis=1).plot(label="Public",color="C1")

plt.xlabel("Trial Year")
plt.ylabel("Average Number of Tests per Variety")
plt.axhline(d[d.location!="region"].groupby(['year'])['variety_id'].value_counts().mean(),
            color="black",ls=":",label="Sample Average")
plt.ylim(0,25)
plt.legend()
```

```python

matrix_private.mean(axis=1).plot(label="Locations",color="black")
location_avg = matrix_private.mean(axis=1).mean()

plt.axhline(matrix_private.mean(axis=1).mean(),color="black",alpha=.3)

d[~d['public_variety']][d.location!="region"].groupby(["year","variety_id"])\
                        ['region'].nunique().unstack().T.mean().plot(color="black",label="Regions",ls="--")

region_avg = d[~d['public_variety']][d.location!="region"].groupby(["year","variety_id"])\
                        ['region'].nunique().unstack().T.mean().mean()

plt.axhline(region_avg,color="black",alpha=.3,ls="--")

plt.legend(frameon=False,ncols=2)
plt.annotate("Location Average: " + str(round(location_avg,2)),xy=(1995,4.7))
plt.annotate("Region Average: " + str(round(region_avg,2)),xy=(1995,2.35))

plt.ylim(0,8)
plt.xlabel("Trial Year")
plt.ylabel("Average Number Per Variety")
plt.savefig("../figs/Location_Region_Avgs.pdf",dpi=300,bbox_inches="tight")
```
